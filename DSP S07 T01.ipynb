### Exercici 3 

Continuant amb les dades de tema esportiu, calcula la correlació de tots els atributs entre sí i representa'ls en una matriu amb diferents colors d'intensitat.

```python
# calculem la matriu de correlació
sport.corr()
```

```python
# representem la matriu de correlació amb matplotlib
corrdf=sport.corr()
plt.figure(figsize=(20, 20))
plt.matshow(corrdf)
plt.show()
```

```python
# la mateixa representació amb seaborn em queda més clarificadora...
plt.figure(figsize=(20, 20))
sns.heatmap(corrdf, annot=True)
plt.show()
```
